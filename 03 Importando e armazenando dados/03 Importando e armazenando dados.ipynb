```python
import os
```

```python
caminho_pasta = "dados/Sales/"

arquivos_lista  = os.listdir(caminho_pasta)

dataframes = []

#read CSV
for arquivo in arquivos_lista:
    if arquivo.endswith(".csv"):

        #criado caminho completo para cada arquivo
        arquivo_completo = os.path.join(caminho_pasta, arquivo)

        #importando cada arquivo para df
        df = pd.read_csv(arquivo_completo)

        #adicionando o df na lista dataframes
        dataframes.append(df)


        #print(arquivo_completo) 

#concatenando todos os dataframes dentro da lista
df_concatenado = pd.concat(dataframes, ignore_index=True)
df_concatenado
```
